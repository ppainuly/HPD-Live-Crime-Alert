```python
from splinter import Browser
from bs4 import BeautifulSoup
import requests
import pymongo
import pandas as pd
from urllib.parse import urlparse
import re
```

# Part 1: Setting up url to scrape

```python
# #Mac Users
# executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
# browser = Browser('chrome', **executable_path, headless=False)

# Windows Users
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://cityofhouston.news/category/incident-updates-hpd/"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup)
```

# Part 2: Click the "articles" that we are wanting to scrape

```python
response_list = soup.find('div', class_ = 'site-content')
print(response_list)
```

# Try it with just one article post, and try to get the address out of it

```python
div_article = response_list.find('article')
print(div_article)
```

```python
article_title = div_article.find("a").text
print(article_title)
```

```python
article_date = div_article.find("span").text
print(article_date)
```

```python
article_p = div_article.find("p").text
print(article_p)
```

```python
print(articleTitles_list)
```

## Now get paragraphs in a list

```python
par_list = []
print(response_list)
button = response_list.find_all('a', class_='button')

for content in button:
    try: 
        button_link = content['href']
        browser.visit(button_link)
        button_html = browser.html
        button_soup = BeautifulSoup(button_html, "html.parser")
        article_list = button_soup.find('article')
        par = article_list.find("div", class_='entry-content')
        par_text = par.find('p').text
        print(par_text)
        print("----------------")
        par_list.append(par_text)
        
    except:
        print("----------------")
```

```python
print(par_list)

```
